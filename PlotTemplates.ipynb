Simple graphing template for matplotlib functions and data plots.
-

Line plots
=

```python
y = np.arange(-5,5,.01) #creating array
E = cosh(y)
P = sinh(y)
```

```python
plot(y,E, color ='red', linewidth = 3, label="function E") #plots function
plot(y,P, color ='blue', linestyle = "--", linewidth = 3, label = "Function P")
title("Line Plots", fontsize= 25)
ax = gca()
ax.spines['right'].set_color('none') #THis gets rid of the edges
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axvline(x=0, color="black") #creates axis lines
axhline(y=0, color="black")
legend(loc="lower right")
xlabel("$x-axis$", fontsize= 15)
ylabel("$y-axis$", fontsize= 15)
grid(True)
show()
```

Subplots
=

```python
subplot(2,1,1)
title("Subplots", fontsize = 25)
plot(y, E, color = "red", linewidth = 3, label= "Function E")
ax = gca()
ax.spines['right'].set_color('none') #THis gets rid of the edges
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axvline(x=0, color="black") #creates axis lines
axhline(y=0, color="black")
xlabel("$x-axis$", fontsize= 15)
ylabel("$y-axis$", fontsize= 15)
grid(True)
subplot(2,1,2)
plot(y,P, color = "blue", linestyle = "--", linewidth = 3, label= "Function P")
ax = gca()
ax.spines['right'].set_color('none') #THis gets rid of the edges
ax.spines['left'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
axvline(x=0, color="black") #creates axis lines
axhline(y=0, color="black")
xlabel("$x-axis$", fontsize= 15)
ylabel("$y-axis$", fontsize= 15)
grid(True)
```
