Electric Potential
=

The dipole potential is just the sum of the potentials from two point charges sepparated by some distance. We'll consider a pair of point charges +- q located at x=0, y=+-d.


$V(x,y) = \frac{kq}{\sqrt{x^2 + (y-d)^2}}-\frac{kq}{\sqrt{x^2 + (y+d)^2}}$


Recall that k = 1/4$\pi\epsilon_0$. To see what is happening we will just let k=1, q=1, d=1. The answer won't depend in an important way on these choices.


```python
#make the x-y grid to cover the plane -2<x<2 and -2<y<2
dx = 0.06
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
X,Y = meshgrid(x,y)
#Make function V(x,y)
k = 1; q = 1; d = 1
V = (k*q/ (X**2 + (Y-d)**2)**(1/2.)) - (k*q/ (X**2 + (Y + d)**2)**(1/2.))
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride=1, linewidth=0, cmap=cm.flag)
cb = fig.colorbar(p,shrink=0.55)
```

```python
#Represent 3D data by contours
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])
contour(X,Y,V,levels, linewidths=4, cmap=cm.flag)
show()
```

##Ex 1. Make a 3-D plot of the function $G(x,y) = e^(-x^2-y^2)$ over the range $-6< (x,y)< 6$ using 51 bins in x and 51 bins in y.

```python
dx = .051
dy = .051
x = arange(-6.,6.,dx)
y = arange(-6.,6.,dy)
X,Y = meshgrid(x,y)
G = e**(-X**2 - Y**2)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, G, rstride = 1, cstride=1, linewidth=0, cmap=cm.seismic)
cb = fig.colorbar(p,shrink=0.5)
cse = ax.contour(X,Y,G, array([0,.2,.4,.6,.8]), zdir ="x", offset=-6)
```
