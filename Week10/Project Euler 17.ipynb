Problem 17
-
If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

Test Case
-

one +two +three +four + five + six = 22 letters

```python
count =0
for n in range(1,7): # start at one end at six
 #start in the hundreds column
    if n % 900 != n:
        count += 14    #nine hundred and
        x = n % 900
    elif n % 800 != n:
        count += 15    #eight hundred and
        x = n % 800
    elif n % 700 != n:
        count += 15    #seven hundred and
        x = n % 700
    elif n % 600 != n:
        count += 13    #six hundred and
        x = n % 600
    elif n % 500 != n:
        count += 14    #five hundred and
        x = n % 500
    elif n % 400 != n:
        count += 14    #four hundred and
        x = n % 400
    elif n % 300 != n:
       count += 15     #three hundred and
       x = n % 300
    elif n % 200 != n:
        count += 13    #two hundred and
        x = n % 200
    elif n % 100 != n:
        count += 13    #one hundred and
        x = n % 100
    else:
        x = n          #n < 100
  # Next the tens column
    if x == 0:
        count -= 3     #remove "and"
    elif x % 90 != x:
        count += 6     #ninety
        x = x % 90
    elif x % 80 != x:
        count += 6     #eighty
        x = x % 80
    elif x % 70 != x:
        count += 7     #seventy
        x = x % 70
    elif x % 60 != x:
        count += 5     #sixty
        x = x % 60
    elif x % 50 != x:
       count += 5      #fifty
       x = x % 50
    elif x % 40 != x:
        count += 5     #forty
        x = x % 40
    elif x % 30 != x:
        count += 6     #thirty
        x = x % 30
    elif x % 20 != x:
        count += 6     #twenty
        x = x % 20
    elif x % 10 != x:
       if x == 17:     #numbers between ten and nineteen
           count += 9
       elif x == 11 or x == 12:
           count += 6
       elif x == 15 or x == 16:
           count += 7
       elif x == 10:
           count += 3
       else:
           count += 8
       x = 0
  #lastly the ones column
    if x == 1 or x == 2 or x == 6:   #numbers between one and nine
        count += 3
    elif x == 3 or x == 7 or x== 8:
        count += 5
    elif x == 4 or x == 5 or x == 9:
        count += 4
 
 
print count
```
