3. The Beta(4,3) Distribution
=

a)
-

```python
g = lambda x: 60*x**3*(1-x)**2
x = arange(0,1.01,0.01)
myplot = plot(x, g(x), lw=3)
show()
```

```python
N =10000
xmin = 0
xmax = 1
subx = x[logical_and((x > xmin),(x<xmax))]
ymin = 0
ymax = max(g(subx))*1.05
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);

myplot = plot(x,g(x), lw=3)
xlim(0.,1.0)

plot(samples_x[1:200],samples_y[1:200],'.')
show()
```

```python
newmask = (samples_y < g(samples_x))
sum(newmask)

myplot = plot(x,g(x), lw=3)
xlim(0.,1.0)

plot(samples_x[1:200],samples_y[1:200],'.')
plot(samples_x[newmask[:200]],samples_y[newmask[:200]], 'r.')
show()
```
