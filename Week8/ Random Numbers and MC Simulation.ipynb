Example 1: Analytic Inverse CDF Transform
==

```python
x = arange(0.,pi+0.001,0.001)
f = lambda x: sin(x)/2
F = lambda x: (cos(0)-cos(x))/2
    
plot(x,f(x),label='True Distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#number of samples
N=10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#samples x values
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label='Sampled Distribution')
plot (x,f(x),label='True Distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

Exercise 1. Exponential function - analytic
=

```python
x = arange(0,1000+0.01,0.01)
y = 1
g = lambda x: y*exp(-y*x)
G = lambda x: -exp(-y*x)

#number of samples
N=10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Ginv = lambda z: -log(-y*z)
#samples x values
xsamples = Ginv(z)
```

```python
y=1
plot(x,g(x),label='True Distribution',lw=3)
plot(x,G(x),label='Cumulative function',lw=3)

xlim(0,5.)
ylim(-1.5,1.5)

legend(loc=0)
show()
```

```python
hist(xsamples,50,normed=True,label='Sampled Distribution')
plot (x,g(x),label='True Distribution',lw=3,color='red')

xlim(0,5.)
ylim(-3,3.)

legend(loc=0)
show()
```
