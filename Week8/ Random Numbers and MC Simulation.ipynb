Example 1: Analytic Inverse CDF Transform
==

```python
x = arange(0.,pi+0.001,0.001)
f = lambda x: sin(x)/2
F = lambda x: (cos(0)-cos(x))/2
    
plot(x,f(x),label='True Distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
#number of samples
N=10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#samples x values
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label='Sampled Distribution')
plot (x,f(x),label='True Distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

```python
N= 10000
def distribute1D(x,prob,N):
    sum_dist = np.cumsum(prob)
    normal = sum_dist/sum(prob)
    
    z = zeros(N, float)
    
    for i in range(N):
        z[i] = random.uniform(0,1)
        
    index = zeros(N,int)
    
    for i in range(N):
        index[i] = np.searchsorted(normal,z[i])
        
    value = zeros(N,float)
    
    for i in range(N):
        value[i] = x[index[i]]
        
    return value
```

```python
Prob_Dist = I(x)


Sum_Prob_Dist = np.cumsum(Prob_Dist)
Norm_Prob_Dist = Sum_Prob_Dist/sum(Prob_Dist)

photons = distribute1D(x,Prob_Dist,10000)

```
