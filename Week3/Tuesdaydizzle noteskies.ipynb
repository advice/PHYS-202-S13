```python
#Simple Line Plot
t = arange(0.0, 1.01, .01) #stands for "a range" not arrange
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
pi = 3.0 # Dope
print pi
pi = np.pi
print pi
```

```python
plot(t,s, linewidth = 3, label = "$sin(4{\pi}t)$")
plot(t,c, color = "red", linewidth = 3, label = "$cos(4{\pi}t)$")
xlabel("Time (s)", fontsize = 15)
ylabel("Voltage (mV)", fontsize = 15)
title("Meg Crash Plot, Yo", fontsize = 25)
legend()
grid(True)
savefig("simple_plot.pdf")
show()
```
